```python
from mplsoccer import Pitch, Sbopen
import matplotlib.pyplot as plt

# read data
parser = Sbopen()
df, related, freeze, tactics = parser.event(7478)
df1 = df[(df.team_name == 'OL Reign') &  (df.type_name == 'Shot') ].copy()
df2 = df[(df.team_name == 'Houston Dash') & (df.type_name == 'Shot') ].copy()


pitch = Pitch()
fig, ax = pitch.draw(figsize=(8, 6))

hull = pitch.convexhull(df1.x, df1.y)
poly = pitch.polygon(hull, ax=ax, edgecolor='black', facecolor='cornflowerblue', alpha=0.3)
scatter = pitch.scatter(df1.x, df1.y, ax=ax, edgecolor='black', facecolor='cornflowerblue',label='OL Reign')


hull = pitch.convexhull(df2.x, df2.y)
poly = pitch.polygon(hull, ax=ax, edgecolor='black', facecolor='red', alpha=0.3)
scatter = pitch.scatter(df2.x, df2.y, ax=ax, edgecolor='black', facecolor='red',label='Houston Dash')
plt.title('Disparos a porteria')
plt.legend()
plt.show()


df1 = df[(df.team_name == 'OL Reign') &  (df.type_name == 'Interception') ].copy()
df2 = df[(df.team_name == 'Houston Dash') & (df.type_name == 'Interception') ].copy()

pitch = Pitch()
fig, ax = pitch.draw(figsize=(8, 6))

hull = pitch.convexhull(df1.x, df1.y)
poly = pitch.polygon(hull, ax=ax, edgecolor='black', facecolor='cornflowerblue', alpha=0.3)
scatter = pitch.scatter(df1.x, df1.y, ax=ax, edgecolor='black', facecolor='cornflowerblue', label='OL Reign')


hull = pitch.convexhull(df2.x, df2.y)
poly = pitch.polygon(hull, ax=ax, edgecolor='black', facecolor='red', alpha=0.3)
scatter = pitch.scatter(df2.x, df2.y, ax=ax, edgecolor='black', facecolor='red',label='Houston Dash')
plt.title('Intercepciones de pases')
plt.legend()
plt.show()
```
